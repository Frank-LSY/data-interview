```python
# 得到2011-2020年数据并整理
def pd_get_data(url,outfile):
    tables=pd.read_html(url)
    dta = tables[0]
    print(outfile[:6],end='\t')
    if outfile[5]=='学': 
        df = dta.loc[2:18,[0,1,2,3,4,5,6]]
    else:
        df = dta.loc[2:17,[0,1,2,3,4,5,6]]
    column_names = {
        0:'科目',1:'总分A',2:'单科A-100',3:'单科A-150',
        4:'总分B',5:'单科B-100',6:'单科B-150'
    }
    df = df.rename(columns=column_names).set_index('科目')
    ori = 'data/{}'.format(outfile)
    df.to_csv(ori)
    df
```
